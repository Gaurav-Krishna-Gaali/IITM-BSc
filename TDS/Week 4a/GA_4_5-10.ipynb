# QUESTION 9 AND 10
Predict value of currency "CHF" from currency"EUR" where the "EUR" value is the average value of the currency "EUR"


```python
new_data = data[["EUR" , "CHF"]]
new_data
```

```python
X = np.array( new_data.iloc[:,0] ).reshape(-1,1)
Y = np.array( new_data.iloc[:,1] ).reshape(-1,1)

r = np.corrcoef (new_data.iloc[:,0] , new_data.iloc[:,1]) [0][1]
model = LinearRegression().fit(X , Y)
c = model.intercept_
m = r*(np.std(Y) / np.std(X))
x = np.average(X)

predict_value = m*x + c
```

```python
predict_value
```

# Question 7 and 8
Example : Find the regression slope for the currency pair- "EUR-HKD"


```python
new_data2 = data[["EUR" , "HKD"]]
```

```python
x = np.array( new_data2.iloc[:,0] ).reshape(-1,1)
y = np.array( new_data2.iloc[:,1] ).reshape(-1,1)

n = np.size(x)

x_mean = np.mean(x)
y_mean = np.mean(y)
x_mean,y_mean
Sxy = np.sum(x*y)- n*x_mean*y_mean
Sxx = np.sum(x*x)-n*x_mean*x_mean

b1 = Sxy/Sxx
b0 = y_mean-b1*x_mean
print('slope b1 is', b1)
print('intercept b0 is', b0)
```
